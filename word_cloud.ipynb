```python
# import libraries
import pandas as pd
import numpy as np
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
import re
import matplotlib.pyplot as plt
%matplotlib inline
import math
import warnings
warnings.filterwarnings('ignore')
```

```python
# Read in the train_rev1 datafile downloaded from kaggle
df = pd.read_csv(r'C:\Users\不知道叫什么\Desktop\HKU sme2\ARIN7102 Applied data mining and text analytics\project/job_description.csv')
df.head(2)
```

```python
Engineering=df[df['Category'] == 'Engineering Jobs']
Social = df[df['Category'] == 'Social work Jobs']
```

```python
from wordcloud import WordCloud
import jieba
import re
```

## Social

```python
Social_text =' '.join(Social['FullDescription'])
clean_text = re.sub(r'[^\w\s]', '', Social_text)
clean_text = clean_text.lower()

nltk.download('stopwords')
nltk.download('punkt')

stop_words = set(stopwords.words('english'))
words = word_tokenize(clean_text)
clean_words = [word for word in words if word.lower() not in stop_words]
```

```python
from collections import Counter
word_freq = Counter(clean_words)
```

```python
filtered_words = {}
for word,freq in word_freq.items():
    if word.strip() and not word.isspace():
        filtered_words[word] = freq
        
wordcloud = WordCloud(font_path="simhei.ttf", background_color="white", width=800, height=600)
wordcloud.generate_from_frequencies(filtered_words)

plt.figure(figsize=(8, 8), facecolor=None)
plt.imshow(wordcloud)
plt.axis("off")
plt.tight_layout(pad=0)
plt.show()
```
